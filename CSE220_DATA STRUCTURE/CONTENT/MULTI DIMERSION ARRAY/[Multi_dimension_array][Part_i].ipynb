# Interaction of 2D matrix

```python
arr= np.array([[1,2,3,4], [9,91,99,98]])

row_len, col_len= arr.shape
print(arr)
print("row : ",row_len)
print("col : ", col_len)
```

```python
for r in range(row_len):
  for c in range(col_len):
    print(arr[r][c],end=" ")
  print()
```

# Scalar and  Matrix multiplication  

```python
A= np.array([[1,2,3,4], [11,22,33,44]])
print("the matrix : ")
print(A)

# Now we will multiply our matrix with 2
print()
print(" multliplying 2 with the matrix  : ")


row,col= A.shape
for r in range(row):
  for c in range(col):
    A[r][c]= 2 * A[r][c]
print( "after the matrix A ")
print()
print(A)
```

```python
A=np.array( [[0,5],[1,2],[3,4]] )
B=np.array( [[3,2,1],[1,2,4]] )
print("A(3,2):")
print(A)
print("B(2,3):")
print(B)
Ar_len , Ac_len = A.shape[0], A.shape[1]
Br_len , Bc_len = B.shape[0], B.shape[1]
#Here, Bc_len==1 and Ac_len==Br_len
C=np.zeros( shape=(Ar_len,Bc_len), dtype=int)
Cr_len, Cc_len = C.shape[0], C.shape[1]
for i in range(Cr_len):
    for j in range(Cc_len):
        sum=0
        for k in range(Br_len):
           sum += A[i][k] * B[k][j]
        C[i][j] = sum
print("A X B = C(3,3):")
print(C)
```
